  <h2>🔴دسته‌بندی ویژگی‌ها در دیتاست</h2>
  <font size="4">
      ویژگی‌های این دیتاست مربوط به بررسی بافت و ساختار هسته‌ی سلول‌های سینه در تصاویر میکروسکوپی هستند. این ویژگی‌ها از نظر آماری و هندسی در سه گروه زیر طبقه‌بندی شده‌اند که هر گروه می‌تواند در تشخیص خوش‌خیم یا بدخیم بودن توده‌ها نقش کلیدی داشته باشد.<br>
      <li>
        <b>ویژگی‌های میانگین (Mean Features):</b>
        <br>
        این گروه از ویژگی‌ها میانگین اندازه‌گیری‌های مربوط به هر نمونه را توصیف می‌کنند. مثلاً <code>radius_mean</code> میانگین شعاع هسته‌هاست و <code>texture_mean</code> میانگین بافت تصویر است. این مقادیر نمای کلی از ساختار فیزیکی توده را ارائه می‌دهند.
        در بحث‌های هندسی، این ویژگی‌ها معمولاً با مشخصه‌های اصلی شکل هندسی (مساحت، محیط، شعاع، یکنواختی، پیچیدگی مرز و ...) مرتبط هستند.
      </li>
      <br>
      <li>
        <b>ویژگی‌های خطای استاندارد (Standard Error Features):</b>
        <br>
        این دسته نشان‌دهنده‌ی میزان نوسان یا پراکندگی اندازه‌گیری‌ها برای هر ویژگی هستند. مثلاً <code>radius_se</code> مقدار انحراف استاندارد اندازه شعاع‌ها در نقاط مختلف سلول است.
        این ویژگی‌ها به شما اطلاعاتی در مورد <u>پایداری</u> ساختار یا تغییرپذیری شکل سلول می‌دهند. در پردازش هندسی، نوسان بالا ممکن است نشانگر لبه‌های نامنظم یا تغییرات غیرنرمال در بافت باشد.
      </li>
      <br>
      <li>
        <b>ویژگی‌های بزرگ‌ترین مقدار (Worst / Largest Features):</b>
        <br>
        این گروه نماینده‌ی بیشترین مقدار مشاهده‌شده برای هر ویژگی است (مثلاً <code>radius_worst</code> بزرگ‌ترین شعاع مشاهده‌شده در مرز سلول).
        این مقادیر معمولاً نشان‌دهنده‌ی حداکثر انحراف از حالت عادی هستند و در طبقه‌بندی توده‌های بدخیم که تمایل به رشد نامنظم دارند، بسیار مهم‌اند.
        در هندسه‌ی محاسباتی، این ویژگی‌ها می‌توانند نمایانگر نقاطی با بیش‌ترین تحدب، پیچیدگی یا بی‌نظمی باشند.
      </li>
      <br><br>
      <font size=4 color='red'>جمع‌بندی</font>: در فرآیند Feature Engineering، درک دقیق این سه دسته ویژگی به شما کمک می‌کند تا بتوانید با ترکیب آن‌ها، ویژگی‌های جدید هندسی بسازید (مانند نسبت حداکثر به میانگین، ضریب نوسان، نسبت سطح به محیط، و غیره) که به مدل در تشخیص دقیق‌تر کمک خواهند کرد.

    <h2><li>راهنمای تکمیل تابع <code>explore_dataset</code></li></h2>
        <font size=4>
        در این تمرین، هدف بررسی مقدماتی کیفیت و ساختار دیتاست با استفاده از یک تابع پایتونی به نام <code>explore_dataset</code> است. این تابع باید اطلاعات پایه‌ای راجع به دیتافریم ورودی را نمایش دهد تا آماده‌سازی داده‌ها برای مراحل بعدی تحلیل یا مدل‌سازی راحت‌تر شود.
    <h3><strong>هدف اصلی تابع:</strong></h3>
        <li>نمایش خلاصه‌ای از اطلاعات دیتافریم (تعداد ستون‌ها، نوع داده‌ها، مقادیر غیر‌خالی)</li>

        <li>بررسی تعداد مقادیر یکتا در هر ستون</li>
        <li>نمایش نوع داده‌ (Data Type) هر ستون</li>
    <h3><strong>آنچه از شما انتظار می‌رود:</strong></h3>
        <li>هر بخش از تابع را با دقت تکمیل کنید تا خروجی‌های واضح و مفیدی تولید کند.</li>
        <li>در صورت تمایل، می‌توانید قسمت‌هایی برای رسم نمودارهای تکمیلی اضافه کنید (اختیاری).</li>
        <li>در پایان اجرا، خروجی باید نمایی خلاصه و قابل فهم از وضعیت فعلی دیتاست ارائه دهد.</li>
    <h3><strong>نکته پایانی:</strong></h3>
        <font size=4>
        این نوع توابع برای هر پروژه‌ی داده‌کاوی یا یادگیری ماشین به‌عنوان یک گام ابتدایی ضروری هستند. با تسلط بر این بررسی‌ها، شما درک عمیق‌تری از داده‌های خام پیدا خواهید کرد و تصمیمات بهتری در مراحل بعدی می‌گیرید.

```python
# TODO : تابع زیر را کامل کنید
def explore_dataset(df):
    """
    انجام بررسی اولیه روی یک دیتافریم pandas.

    پارامترها
    ----------
    df : pd.DataFrame
        دیتافریم ورودی که باید بررسی شود.

    name : str, اختیاری (پیش‌فرض: "Dataset")
        نام دیتاست برای نمایش در خروجی‌ها.

    عملکرد تابع
    ------------
    - نمایش اطلاعات کلی دیتافریم با استفاده از df.info().
    - نمایش تعداد داده‌های گم‌شده در هر ستون.
    - گزارش تعداد ردیف‌های تکراری.
    - نمایش تعداد مقادیر یکتا برای هر ستون.
    - نمایش نوع داده‌ای (dtype) تمام ستون‌ها.
    """
    print('نمایش کلی دیتافریم :')
    print(df.info())
    print(100 * '*')

    print(f'head data frame : \n{df.head(5)}')
    print(100 * '*')

    print(f'shape data frame : {df.shape}')
    print(100 * '*')

    print(f'columns data frame : {df.columns}')
    print(100 * '*')

    print(f'type data frame : \n{df.dtypes}')
    print(100 * '*')

    print(f'null data frame :\n{df.isnull().sum()}')
    print(100 * '*')

    print(f'duplicated : \n{df.duplicated().sum()}')
    print(100 * '*')

    print(f'diagnosis(target) value count :{df.value_counts('diagnosis')}')
    print(100 * '*')
```

```python
explore_dataset(df)
```

```python
def plot_numerical_histograms(dataframe):
    # جدا کردن ستون هدف از ویژگی‌ها
    if 'diagnosis' in df.columns:
        features = dataframe.drop('diagnosis', axis=1)
    else:
        features = dataframe # اگر target وجود نداشت، همه ستون‌ها را ویژگی در نظر بگیر

    # تعیین تعداد ستون‌ها و ردیف‌های مورد نیاز برای نمودارها
    num_features = features.shape[1]
    num_cols = 4  # تعداد ستون‌ها در گرید نمودار
    num_rows = (num_features + num_cols - 1) // num_cols # محاسبه تعداد ردیف‌ها

    plt.figure(figsize=(num_cols * 5, num_rows * 4)) # تنظیم اندازه کلی شکل

    for i, col in enumerate(features.columns):
        plt.subplot(num_rows, num_cols, i + 1)
        sns.histplot(data=dataframe, x=col, kde=True, palette='viridis') # kde=True برای نمایش منحنی تخمین چگالی
        plt.title(f'توزیع {col}')
        plt.xlabel(col)
        plt.ylabel('فراوانی')

    plt.tight_layout() # تنظیم فاصله‌ها برای جلوگیری از همپوشانی
    plt.show()

# فراخوانی تابع برای رسم هیستوگرام‌ها
plot_numerical_histograms(df)
```

```python
# TODO: طراحی شبکه عصبی
def build_model(input_dim, l2_lambda=0.001):
    """
    مدل ANN ساده با مقدار متوسطی از dropout و regularization (L2).
    """
    initializer = HeNormal()
    regularizer = l2(l2_lambda)

    model=Sequential([
        #  لایه ورودی و مخفی اول
            Dense(units=32,activation='relu',input_shape=(input_dim,),
                  kernel_initializer=initializer, kernel_regularizer=regularizer),
                #  لایه مخفی دوم
            Dense(units=64, activation='relu',
                  kernel_initializer=initializer, kernel_regularizer=regularizer ),
                # لایه خروجی
            Dense(units=1, activation='sigmoid', kernel_initializer=initializer)
                    ])



    model.compile(optimizer=Adam(learning_rate=0.0001),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model
    
```
